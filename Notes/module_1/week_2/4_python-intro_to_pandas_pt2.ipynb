### 2.2. Subestting

```python
df.loc[]
```

```python
df["banana_density"].lt(.5)
```

```python
df["apple_density"].lt(.5)
```

```python
df["banana_density"].lt(.5) | df["apple_density"].lt(.5)
```

```python
df["banana_density"].lt(.5) & df["apple_density"].lt(.5)
```

```python
df.loc[df["banana_density"].lt(.5) | df["apple_density"].lt(.5), ["banana_density", "apple_density"]]
```

```python
df.loc[df["banana_density"].lt(.5) & df["apple_density"].lt(.5), ["banana_density", "apple_density"]]
```
