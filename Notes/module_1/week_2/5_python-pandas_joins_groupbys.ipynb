Replicate the following SQL query in pandas:

```sql
SELECT
    oi.order_id,
    oi.seller_id,
    oi.product_id,
    p.product_category_name
FROM olist.order_items oi
    INNER JOIN olist.products p
    ON oi.product_id = p.product_id
```

````python
order_items = pd.read_sql("SELECT * FROM order_items;", conn)
product = pd.read_sql("SELECT * FROM products;", conn)
````

Do a join

````python
order_items_products = order_items.merge(right=product,
                                         how="inner",
                                         on="product_id")
````

````python
order_items_products.head()
````

````python
order_items_products[["order_id",
                      "product_id",
                      "seller_id",
                      "product_category_name"]]
````

````python
order_items_product_category = order_items.merge(right=product,
                                                 how="inner",
                                                 on="product_id")[["order_id",
                                                                   "product_id",
                                                                   "seller_id",
                                                                   "product_category_name"]]
````

Add the group by

```sql
SELECT
    p.product_category_name,
    COUNT(*)                 AS items_sold
FROM olist.order_items oi
    INNER JOIN olist.products p
    ON oi.product_id = p.product_id
GROUP BY p.product_category_name
```

````python
order_items_product_category[["product_category_name", "order_id"]].groupby("product_category_name").count()
````

Multiple ways to subset as the count function applies to all columns

````python
order_items_product_category.groupby("product_category_name").count()[["order_id"]]
````

````python
sales_per_cat = (order_items_product_category
                 .groupby("product_category_name")[["order_id"]]
                 .count()
                 .reset_index()
                 .rename(columns={"order_id": "items_sold"}))
````

````python
sales_per_cat
````

Do everything at once

````python
(order_items
 .merge(right=product,
        on="product_id")[["product_category_name", "product_id"]]
 .groupby("product_category_name")
 .count()
 .reset_index()
 .rename(columns={"product_id": "items_sold"})
 .sort_values("items_sold", ascending=False)
 .reset_index(drop=True))
````

A few SQL tricks. Different ways to pass sql queries

````python
# 1. indicate new lines with backslash
items_per_cat_sql = pd.read_sql("SELECT \
                                    p.product_category_name, \
                                    COUNT(*)                 AS items_sold \
                                FROM olist.order_items oi \
                                    INNER JOIN olist.products p \
                                    ON oi.product_id = p.product_id \
                                GROUP BY p.product_category_name \
                                ORDER BY 2 DESC", 
                                conn)
````

````python
# 2. use triple quotes
items_per_cat_sql = pd.read_sql("""SELECT
                                    p.product_category_name,
                                    COUNT(*)                 AS items_sold
                                FROM olist.order_items oi
                                    INNER JOIN olist.products p
                                    ON oi.product_id = p.product_id
                                GROUP BY p.product_category_name
                                ORDER BY 2 DESC""", 
                                conn)
````

````python
# 3. isolate query in dedicated code cell and store in variable
query = """SELECT
            p.product_category_name,
            COUNT(*)                 AS items_sold
        FROM olist.order_items oi
            INNER JOIN olist.products p
            ON oi.product_id = p.product_id
        GROUP BY p.product_category_name
        ORDER BY 2 DESC"""
````

````python
items_per_cat_sql = pd.read_sql(query, conn)
````

````python
# 4. isolate query in dedicated module and import
import sql.queries as q
items_per_cat_sql = pd.read_sql(q.query, conn)
````

### Decoding values in a column

First approach: Explicit overwriting from cheap to expensive

````python
# 1. create new empty column
price_categories = order_items[["price"]].assign(price_category=np.nan)
````

````python
price_categories["price"].lt(150)
````

````python
# 2. overwrite rows where price is cheap
price_categories.loc[price_categories["price"].lt(150),["price_category"]] = "cheap"
````

````python
price_categories
````

````python
# 3. Overwrite rows where price is affordable
price_categories.loc[price_categories["price"].between(150,299.99,inclusive=True),["price_category"]] = "affordable"
````

````python
# 4. Overwrite rows where price is expensive
price_categories
````

````python
price_categories["price_category"] = price_categories["price_category"].fillna("expensive")
````

````python
price_categories
````
