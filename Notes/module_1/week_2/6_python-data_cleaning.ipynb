# 1. Missing values

```python
# df.isna().sum()[df.isna().sum().gt(0)]

# assign output to a variable
null_columns = df.isna().sum()
null_columns[null_columns.gt(0)]
```

```python
df.loc[df["cylinders"].isna(), "cylinders"].to_frame()
```

```python
null_columns[null_columns.gt(0)] / len(df)
```

We might as well drop columns that have a proportion of >50% missing values

```python
null_prop = null_columns[null_columns.gt(0)] / len(df)
null_prop
```

```python
null_prop[null_prop.gt(.5)]
```

```python
list(null_prop[null_prop.gt(.5)].index)
```

```python
df = df.drop(columns=null_prop[null_prop.gt(.5)].index)
```

```python
(df.isna().sum() / len(df))[(df.isna().sum() / len(df)).gt(0)]
```

```python
def checkna(df):
    return (df.isna().sum() / len(df))[(df.isna().sum() / len(df)).gt(0)]
```

```python
df = df.drop(columns=["eng_dscr"])
```

```python
checkna(df)
```

We want to further investigate the missing values in cylinders

```python
important_columns = ["year", "make", "model", "trany", "drive", "fuelType", "cylinders", "displ"]
```

```python
df.loc[df["cylinders"].isna(), important_columns]
```

Makes sense! Electric cars don't have cylinders. Fill missing values with 0

```python
df[["cylinders","displ"]] = df[["cylinders","displ"]].fillna(0)
```

```python
checkna(df)
```

```python
# create a function that takes a dataframe, a column label for that dataframe and a precentile threshold as an input
# and returns whether the min of that column is equal to the percentile

def check_low_information(df, col, perc):
    min_value = df[col].min()
    perc_value = np.percentile(df[col], perc)
    return min_value == perc_value
```

```python
check_low_information(df, "barrelsA08", 90)
```
