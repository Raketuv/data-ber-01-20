### 1. Get rank and school name

```python
school_rank = str(school_list_item.h3.a.contents[0])
```

```python
school_rank
```

```python
# Get Rank
rank_pattern = r"^(\d+)\."
rank = int(re.findall(rank_pattern, school_rank)[0])
```

```python
rank
```

```python
# Get School name
name_pattern = r"^\d+\.\s(.*)$"
school = re.findall(name_pattern, school_rank)[0]
```

```python
# Alternatively, split string
school_rank.split(".")
```

```python
print(school)
print(rank)
```

### 2. Overall Rating

```python
rating_string = (school_list_item
                 .find("span", class_="longform-rating-text")
                 .contents)[0]

rating_pattern = r"\((.*)\)"
rating = float(re.findall(rating_pattern, rating_string)[0])
```

```python
rating
```

### 3. Stars

```python
# 1. get first star
(school_list_item
 .find("div", class_="ratings title-rating")
 .find_all("span"))
```

```python
# 2. get the class name from one span tag that has a star
first_star_icon = (school_list_item
                   .find("div", class_="ratings title-rating")
                   .find_all("span"))[1]["class"][0]
```

```python
first_star_icon
```

```python
# 3. translate class names to numbers
stars_dict = {"icon-full_star": 1,
              "icon-half_star": .5}

print(stars_dict[first_star_icon])
type(stars_dict[first_star_icon])
```

```python
# 4. do it for all stars

# get a list of all star span tags
star_tag_list = (school_list_item
                 .find("div", class_="ratings title-rating")
                 .find_all("span"))[1:]

# iterate over all tags and translate them
for star_tag in star_tag_list:
    star = star_tag["class"][0]
    print(stars_dict[star])
```

```python
star_tag_list
```

```python
star_tag_list[0]["class"][0]
```

```python
# do it in list comprehension
star_list = [stars_dict[star_tag["class"][0]] for star_tag in star_tag_list]
```

```python
star_list
```

```python
stars = sum(star_list)
```

```python
stars
```
